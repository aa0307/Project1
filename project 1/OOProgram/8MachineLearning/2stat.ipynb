```python
import matplotlib.pyplot as plt
from scipy import stats
x = [5,7,8,7,2,17,2,9,4,11,12,9,6] # car ages
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] # car speed in km/h 
# y = Slope * X + intercept
slope, intercept, r, p, std_err = stats.linregress (x,y)
# r: Correlation coef, p_value, standard error
print('slope', slope); print('intercept', intercept)
# create a function uses slope and intercept and return a new Y 
def func(x):
    return slope * x + intercept
# run each value of the x array via the function to create a MODEL 
model = list(map(func,x))
# the map() executes for each item in an iterable
plt.scatter(x,y) ; plt.plot(x,model); plt.show()
```

r value (correlation) tells how well does the data fit in a linear regression
p value tells model's statistical significance, should not greater than 0.05
Standard deviation errors std_err, it tells how much the dependent variable differs from the regression model

```python
# Calculation by formula
import numpy as np
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,4,2,4,5]
xmean = np.mean(x)
ymean = np.mean(y)
x_xmean = x - xmean
y_ymean = y - ymean
# slope m
sum_top = sum(x_xmean*y_ymean)
sum_bottom = sum(x_xmean*x_xmean)
m = sum_top/sum_bottom
# intercept b
b = ymean - (m * xmean) # y=mx+b; b=y-mx
# create model
y_model = []
# with slope m and intercept b, calculate y for each x
for i in x:
    y_model.append(i*m+b)
print('y_model',y_model)
plt.scatter(x,y) # training dataset
plt.plot(x,y_model) # predicted data set
plt.show()
```

### R Squared Value

```python
# R Squared Value
sum_y_ymean = sum(y_ymean * y_ymean) # sum of (y - ymean) square
array1 = np.array(y_model)
array2 = np.array(ymean)
subtracted_array = np.subtract(array1,array2)
sum_square_array = sum(subtracted_array * subtracted_array)

R_square = sum_square_array/sum_y_ymean
print('R_square',R_square)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([3,4,2,4,5])

# create an instance of a linear regression model and fit it to the data or train the model with the data by using the fit() function
model = LinearRegression().fit(x,y)

# Correlation value by the .score()
r_sq = model.score(x,y)

print("r_sq", r_sq) # r_square value
print("intercept", model.intercept_) # intercept
print("slope", model.coef_) # slope

# Predict a reponse
y_pred = model.predict(x)
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
```
