### Polynomial Regresion

Polynomial Regression is a form of linear regression in which the relationship between the independent variable X and dependent variable Y is modeled as an n-th degree polynomial


```python
import matplotlib.pyplot as plt
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,
     99,99,100]
# create a model to fit the data
import numpy
mymodel = numpy.poly1d(numpy.polyfit(x,y,1))# order 1
myline = numpy.linspace(1,22,100)# 100samples from 1 to 22
plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
# R-Squared range from 0 to 1 shows correlation
print(r2_score(y,mymodel(x)))
```
