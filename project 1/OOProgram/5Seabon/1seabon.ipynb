### sns.regplot

The plot is used to plot data and a linear regression model fit. There are number of mutually exclusive option for estimating the regression model.

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head(3)
# sns.regplot(data=tips, x='total_bill',y='tip',
#             line_kws={'color':'r','alpha':0.5,'lw':2})
# the semi transparent region around the line is the
# confidence interval, removed by setting ci=None
sns.regplot(data=tips, x='total_bill',y='tip',ci=None,
            line_kws={'color':'r','alpha':0.5,'lw':2})
```

```python
# to create an estimated regression line, you will need 
# to use the scipy module
import scipy.stats as stats # pip install scipy
# To fit a linear regression model to your data, need the stats.linregression
# This function takes two variables as imput and returns the
# slope, intercept, rvalue,pvalue,standard error for the model
# you can then use this information to calculate the
# equation of the regression line
slope, intercept, rvalue,pvalue,stderr = stats.linregress(
    tips['total_bill'],tips['tip'])
# rvalue:correlation coef that measures correlation between 2 datasets
# pvalue:probability of null hypothesis is true (>0.05)
print('slope',slope)
print('intercept',intercept)
print('rvalue',rvalue)
```

### Prediction

```python
# to calculate the predicted value of the dependent variables (y)
# for any given valuje of the independent variable (x)
# to do this, you need the model
# y = slope * x + intercept
# to plot the extended regression line
import matplotlib.pyplot as plt
x_values=[10,60]
y_values=[slope*x + intercept for x in x_values]
plt.plot(x_values,y_values,color='red')
```

```python
# to combine the extended regression line
sns.regplot(x='total_bill',y='tip',data=tips)
x_values=[10,60]
y_values=[slope*x + intercept for x in x_values]
plt.plot(x_values,y_values,color='red')
plt.show()
```

### sns.FacetGrid

Multi-plot grid for plotting conditional relationship

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# initializes the gride but does not plot anything on it
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
sns.FacetGrid(tips)
plt.show()
```

```python
# assign column and/or row variables to add more subplots to the figure
sns.FacetGrid(tips,col='time',row='sex')
```

```python
# to draw a plot on every facet
# pass a function and the name of one or more columns
# in the dataframe to FacetGrid.map()
AA = sns.FacetGrid(tips,col='time',row='sex')
AA.map(sns.scatterplot,'total_bill','tip')
```

```python
# can also .map_dataframe()
BB = sns.FacetGrid(tips, col='time',row='sex')
BB.map_dataframe(sns.histplot, x='total_bill')
```

```python
# hue parametr. Setting this will condition the data on another
# variables and make multiple plots in different colors
CC = sns.FacetGrid(tips, col='time',hue='sex')
CC.map_dataframe(sns.scatterplot, x='total_bill',y='tip')
CC.add_legend()
```

```python
# to add horizontal or vertical reference lines on every facet
# use FacetGrid.refline()
DD=sns.FacetGrid(tips, col='time', margin_titles=True)
DD.map_dataframe(sns.scatterplot,x='total_bill',y='tip')
DD.refline(y=tips['tip'].median())
```

```python
# to annotate each facet, use custom function
import matplotlib.pyplot as plt
def annotate(data,**kws):
    n=len(data)
    ax = plt.gca()
    ax.text(.1,.6,f"N={n}",transform=ax.transAxes)
EE = sns.FacetGrid(tips, col='time')
EE.map_dataframe(sns.scatterplot,x='total_bill',y='tip')
EE.map_dataframe(annotate)
```

```python
# If the Order is >1, it uses numpy.polyfit
# to estimate a regression
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('mpg')
# print(df.head(3))
sns.lmplot(data=df, x='horsepower',y='mpg',order=2)
plt.show()
```

```python
# Example Numpy.polyfit
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(12)
x = np.linspace(0,1,25)
y = np.cos(x) + 0.3 * np.random.rand(25)# polynomial equ
p = np.poly1d(np.polyfit(x,y,4))# fit with 4 order
t = np.linspace(0,1,250)
plt.plot(x,y,'o',t,p(t),'-')
plt.show()
# numpy polyfit is a function that is used to fit the data
# within a polynomial function. By setting data,
# it finding the least number of squares (least square error)
# in the function
# that fit in a polynomial equation.
```
