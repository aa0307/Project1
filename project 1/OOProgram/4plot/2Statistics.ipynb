### Statistics

Statistics is science of **collecting and analysing numerical data** 
in large quantities
- t-distrubution, a type of normal distrubution used for smaller
sizea, where the variance in the data is unknown.
- Z distrubution is a special case of the normal distubution with a 
mean of 0 and standard deviation of 1.
- Skewness is a measure of the asymmetry of a distribution(direction)
- Kurtosis is a measure of the out;ier (tailedness) of a distrubution.
- Standard Deviation is a measure of how spread out numbers are, It is 
the square root of the Variance.
- Variance is the average of the squared differences from the Mean. 

### Probability
Probability is a measure of the **likelihood 可能性** of an event to  
occur. In probaility, events are the outcomes of an experiment.  
The probaility that A or B will occur is the SUM of the probability  
of each event, MINUS the probability of the overlap.   
P(A or B) = P(A)+P(B) - P(A and B)  

Probability works fine when dicussing a number of discrete, possible
outcome.  
When there is a continuous range of possible outcomes. Instead of  
probability, prediction in terms of **Probability Density Function**.

### Binomial Distribution

Binomial Distribution is a **Discrete Distribution** (Normal
distribution is a acontinuous distribution), there is no data point
between any two data points.

It is suited for trails where there are 
- number of observations n is fixed and identical
- only two mutually exclusive outcomes
- each observation is independent (probability is constant)

Example: Coin toss or defective items check

```python
# Coin toss
# 10 filps of a coin, what is the possibility of getting a 'head'
# The distribution is defined at seperate set of events
# n - number of trials
# p - probability of occurence of each trial (toss of coin 0.5 each)
# size - the shape of the returned array (no. of tested times)
# output (result of flipping a coin n times)
# is the samples from the distribution where
# each sample is equal to the number of successes over n trails
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
CC=random.binomial(n=10,p=0.5,size=18) 
print(CC)
DD=CC/10
print('probability of getting a "head" in each trial:\n',DD)
sns.histplot(data=CC)
plt.show()
```
