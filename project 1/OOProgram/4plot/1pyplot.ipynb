### Histogram Plot

The ***Histogram*** is used to summarize discrete or continuous data that are meassured on an **interval** scale. It is used to showcase a grapical presentation that represents the data in the form of frequency.  
A stacked chart, you can get an idea of the total counts and the contribution of each region.  

***Line graphs*** are used to track changes over short and long periods of time. When **Smaller** changes exist, line graphs are better to use than Bar chart.  

whereas a ***Bar Chart*** is a graphical representation of data and the information that is used for the **comparison** of two categories.

#### Use case Example
Restaurant manager might build a histogram to determine how many customers come into a restaurant at different times during the day. The number of customers is on the vertical axis, while the time intervals are on the horizontal axis.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
# create dataset
np.random.seed(42)
N_points = 10000
n_bins = 20
# create distrubution
x=np.random.randn(N_points)
legend = ['distribution']
# create histogram
fig,axs = plt.subplots(1,1,figsize=(10,7),tight_layout = True)
# remove axes splines
for s in ['top','bottom','left','right']:
    axs.spines[s].set_visible(False)
# add padding between axes and labels
axs.xaxis.set_tick_params(pad = 50)
axs.yaxis.set_tick_params(pad = 100)
# add x,y gridlines
axs.grid(color='grey',linestyle='-.',linewidth = 0.5,alpha = 0.6)
# create histogram
N, bins, patches = axs.hist(x,bins=n_bins)
# N is the number of counts in each bin of the histogram
# bins is the left hand edge of each bin
# paches is the individual patches used to creat the histogram
print('N',N)
print('bins',bins) 

# set color
# normalize the data to 0,,,,1 for full range of the colormap
fracs = ((N**(1/5))/N.max())
norm = colors.Normalize(fracs.min(),fracs.max())
# loop through objects and set the color of each accordingly
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(legend)
plt.title("Customized Histogram")
plt.show()
```

```python
# Histogram 
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
n_bins = 20
x = np.random.randn(10000,3) # 2D array 10K rows 3 columns
fig,[(ax0,ax1),(ax2,ax3)] = plt.subplots(2,2,figsize = (8,6))
colors = ['green','black','lime']
# ax0
ax0.hist(x,n_bins, histtype = 'bar', color = colors, label = colors)
ax0.set_title('Bar')
ax0.legend(prop = {'size':10})
# ax1
ax1.hist(x, n_bins, histtype= 'barstacked', stacked = True, color = colors)
ax1.set_title('Barstacked')
# ax2
ax2.hist(x, n_bins, histtype = 'step', stacked = True, fill = False, color = colors)
ax2.set_title('Step', y=1, pad=-20, loc='left')
# ax3
x_multi = [np.random.randn(n) for n in [100000,80000,1000]]# 1D with 100K,80K 1K elements
ax3.hist(x_multi, n_bins, histtype= 'stepfilled', color = colors)
ax3.set_title('Stepfilled', y=1, pad=-20, loc='right')
plt.show()
```

### Pie Chart

Pie Charts are best to use when you are trying to compare parts of the a whole.
They do not show changes over time.

```python
# Ax plot of pie chart
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots(figsize=(8,6),subplot_kw=dict(aspect='equal'))
recipe = ['375 g flour','75 g sugar','250 g butter','300 g berries']
data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]
# create autocpt argument
# autocpt enables you to display the precent value using python string format
def func(pct,allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return '{:.1f}%\n({:d} g)'.format(pct,absolute)

wedges, text, autotexts = ax.pie(data,autopct=lambda pct:func(pct,data),
                                 textprops = dict(color='w'))
ax.legend(wedges,ingredients,title='ingredients',
          loc = 'center left',bbox_to_anchor=(1,0,0.5,1))
plt.setp(autotexts, size=8,weight='bold')
ax.set_title("Bakery: A pie")
plt.show()
```

### Heat Map

A heatmap is a graphical representation of data that uses a system of color coding to represent different values

```python
#Heat Map

import numpy as np
import matplotlib.pyplot as plt
data=np.random.random((8,8))
xlabs=["G1","G2","G3","G4","G5","G6","G7","G8"]
ylabs=['A','B','C','D','E','F','G','H']
fig,ax=plt.subplots(figsize=(6,4))
im=ax.imshow(data)
ax.set_xticks(np.arange(len(xlabs)),labels=xlabs)
ax.set_yticks(np.arange(len(ylabs)),labels=ylabs)
# add the values to each cell
for i in range(len(xlabs)):
    for j in range(len(ylabs)):
        text=ax.text(j,i,round(data[i,j],1),
                     ha='center',va='center',
                     color='w')
# add color bar
cbar=ax.figure.colorbar(im,ax=ax)
cbar.ax.set_ylabel('Color bar',rotation=-90,va='bottom')
plt.show()
```

```python
# Heatmap Example
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
vegetables = ['cucumber','tomato','lettuce','asparagus','potato',
              'wheat','barley']
farmers = ['Joe','Jon','Mary','Dav','James','Bio Ltd','Corny Corp']
harvest = np.array([[0.8,2.4,2.5,3.9,0.0,4.0,0.0],
                   [2.4,0.0,4.0,1.0,2.7,0.0,0.0],
                   [1.1,2.4,0.8,4.3,1.9,4.4,0.0],
                   [0.6,0.0,0.3,0.0,3.1,0.0,0.0],
                   [0.7,1.7,0.6,2.6,2.2,6.2,0.0],
                   [1.3,1.2,0.0,0.0,0.0,3.2,5.1],
                   [0.1,2.0,0.0,1.4,0.0,1.9,6.3]])
fig,ax = plt.subplots(figsize=(10,8))
im = ax.imshow(harvest)
# show all ticks and label them with the repective list entries
ax.set_xticks(np.arange(len(farmers)),labels=farmers)
ax.set_yticks(np.arange(len(farmers)),labels=farmers)
# rotate the tick labels and set their alignment
plt.setp(ax.get_xticklabels(),rotation=45,ha='right',
         rotation_mode='anchor')
# loop over data dimensions and create text annotations
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j,i,harvest[i,j],
                       ha = 'center',va = 'center',color = 'w')
ax.set_title('Harvest of local farmer (in ton/year)')
fig.colorbar(im,ax=ax,label='harvest [t/year]')
fig.tight_layout()
plt.show()
```
