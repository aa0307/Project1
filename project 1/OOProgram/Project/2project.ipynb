```python
# example of .merge()
import pandas as pd

AAA={'Name':['jon','mary','dev'],'Country':['hk','uk','us'],'Role':['ceo','cto','cfo']}
ADF=pd.DataFrame(AAA)
print('ADF',ADF)
BDF=pd.DataFrame({'ID':[1,2,3],'Name':['jon','ann','leo']})
print('BDF',BDF)
CM1=ADF.merge(BDF)
print('merged\n',CM1)
CM2=ADF.merge(BDF,on='Name') # specific column
print('CM2\n',CM2)
```
