## <span style="color: yellow;"> Remove Empty Cell (ie NaN)</span>
.dropna

```python
#remove rows that contain empty cells, .dropna()

import pandas as pd
path="C:/Users/User/Desktop/project 1/VS/data_folder/"
file_name="data1.xlsx"
df = pd.read_excel(f"{path}{file_name}")

drop_empty_df = df.dropna()

# Pandas distionction between str and object types, object can be number
# to_string() directly transfer the pd.DataFrame object to a string object,
# print no display limit
print(drop_empty_df.to_string()) #row 18, 22, 28 NaN removed

#dropna() returns a new DataFrame, and will not change the original.
#Replace original uses dropna(inplace = True)
```


## <span style="color: yellow;"> Replace empty cell with a new value</span>

.fillna

```python
#insert a new value, fillna()

import pandas as pd
path="C:/Users/User/Desktop/project 1/VS/data_folder/"
file_name="data1.xlsx"
df = pd.read_excel(f"{path}{file_name}")

df.fillna(130, inplace = True) #replace all empty cells with a value (130)
print(df.to_string())
```

## <span style="color: yellow;"> Replace NULL values only in the "income" columns</span>


```python
#Replace NULL values only in the "Calories" column with the number 130:

import pandas as pd
path="C:/Users/User/Desktop/project 1/VS/data_folder/"
file_name="data1.xlsx"
df = pd.read_excel(f"{path}{file_name}")

df['Incomes'].fillna(130, inplace = True) #replace only the empty cells 18,28 in "Incomes" (space count) with a 130
print(df.to_string())
```

## <span style="color: yellow;"> Calculate the MEAN</span>
.mean

```python
#Calculate the MEAN, and replace any empty values with it:

import pandas as pd
path="C:/Users/User/Desktop/project 1/VS/data_folder/"
file_name="data1.xlsx"
df = pd.read_excel(f"{path}{file_name}")

x = df["Incomes"].mean() # get the mean value
print('mean value: ',x)
newx_df=df["Incomes"].fillna(x) # fill the space with the mean value
print(newx_df.to_string())
```

## <span style="color: yellow;"> MEDIAN</span>

```python
#Calculate the MEDIAN, Median = the value in the middle,
# and replace an empty values with it:

import pandas as pd
path="C:/Users/User/Desktop/project 1/VS/data_folder/"
file_name="data1.xlsx"
df = pd.read_excel(f"{path}{file_name}")
print(type(df))

y = df["Incomes"].median()
print('median: ',y)
newy_df=df["Incomes"].fillna(y)
print(newy_df.to_string())
```

## <span style="color: yellow;"> Calculate the MODE</span>
.mode

```python
#Calculate Mode = the value that appears most frequently.

import pandas as pd
path="C:/Users/User/Desktop/project 1/VS/data_folder/"
file_name="data1.xlsx"
df = pd.read_excel(f"{path}{file_name}")

z = df["Incomes"].mode()[0] # mode([0]) return single value
print('mode: ',z)
df["Incomes"].fillna(z, inplace = True)
#print(df.to_string())
#print(df.loc[0])
print(df.tail(10))
```
