Replace values where the condition is False

```python
# Let's assume a species is large if its SepalLength is bigger 
# than mean. Create a new column 'LargeSpecies' and fill with 
# Boolean to identify if the species belong to LargeSpecies.

# You may create a column to fill with False first.
# $> df.loc[df.index, 'LargeSpecies'] = False

# reminder
# Don’t confuse Numpy.where with
# Pandas DF.where and Series.where
```

```python
df['SepalLength'].mean()
```

```python
df.loc[df.index, 'LargeSpecies'] = False
df.sample(2)
```

```python
df['LargeSpecies'] = df['LargeSpecies'].where(
    df['SepalLength']<df['SepalLength'].mean(),True)
df.sample(5)
```

Replace values where the condition is True

```python
# df.mask() is the oppsite of df.where()
# df.mask(cond, other) will replace value with other if cond is True.

# Let's create another column
# df.loc[df.index, 'SmallSpecies'] = False
# Replace the Boolean to True if 'SepalLength' smaller and equal 
# to mean.
```

```python
df.loc[df.index, 'SmallSpecies'] = False
```

```python
df['SmallSpecies'] = df['SmallSpecies'].mask(
    df['SepalLength'] <= df['SepalLength'].mean(), True)
df.sample(5)
```

```python
# Double check. Add up thr LargeSpecies(True) and Small Species(True)
df.loc[df['LargeSpecies']==True].shape[0] + df.loc[df['SmallSpecies']==True].shape[0]
```

Fill value with numpy.where()

Requirement:
1. Create new column 'SampleSize'
2. Fill it 'Large' if SepalLength > mean, else 'Small'.

Given Input:
Drop column 'LargeSpecies' and 'SmallSpecies'
df = df.drop(['LargeSpecies','SmallSpecies'], axis=1)
Desired Output:

```python
df = df.drop(['LargeSpecies','SmallSpecies'], axis=1)
df.sample(2)
```

```python
df['SampleSize'] = np.where(df['SepalLength']>df['SepalLength'].mean(), 'Large', 'Small')
df['SampleSize'].value_counts()
```

```python
df.sample(3)
```

Change DataFrame column postion

```python
df = df[['Species', 'SepalLength', 'SepalWidth',
         'PetalLength', 'PetalWidth', 'SampleSize']]
df.sample(3)
```

Apply a function on a column

Let's change the data in SampleSize field to Uppercase

```python
df['SampleSize'] = df['SampleSize'].apply(lambda x: x.upper())
df.sample(5)
```

Map dict or function on a column

```python
df['SampleSize'] = df['SampleSize'].map({'SMALL': 'Small Iris',
                                         'LARGE': 'Large Iris'})
df.sample(5)
```

Applymap on multiple column or whole DF

Increase floating point in all numeric column

```python
df[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']] = df[[
    'SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']].applymap(lambda x: f'{x:.3f}')
df.sample(5)




```
