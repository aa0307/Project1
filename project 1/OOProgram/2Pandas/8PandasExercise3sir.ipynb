```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
hsi=yf.download('^HSI',start='2020-01-01',end='2023-01-01')
#hsi=yf.download('0005.hk',start='2020-01-01',end='2023-01-01',interval='60m')
hsi.head(3)
```

```python
# 62
# Moving Average - rolling mean - moving average 10 days mean value

hsi['MA10'] = hsi['Adj Close'].rolling(10).mean()
hsi.tail(5)
```

```python
# 63
# Exponentially Weighted Moving Averag (EWMA)
# EWMA is a quantitative technique used as a forecasting model for time series analyis
# The concept of using a moving average is designed to give more weight 
# for newer data points
# and less for older ones. The weight decline exponentially
# as the data points get older 
# span=10 is calculated using the last 10 observation
hsi['EMA10']=hsi['Adj Close'].ewm(span=10).mean()
hsi.tail(3)
```

```python
# 64
# slicing series to get the last row of data
print(hsi['Adj Close'][-10:])
```

```python
hsi['Adj Close'][-10:].rolling(10).mean()[-1]
```

```python
hsi['Adj Close'][-10:].ewm(span=10).mean()[-1]
```

```python
# 65
# Rolling Kurtosis
# Kurtosis indicates how much data resides in the tails
# for investors, high kurtosis of the return distribution curve implies that
# there have been many price fluctuation in the past
# away from the average returns for the investment
# so, an investor might experience extreme price fluctuations with an
# investment with high kurtosis.
hsi['Kurt50'] = hsi['Adj Close'].rolling(50).kurt()
hsi.tail(3)
```

```python
# 66 Skewness
# Skewness gives the direction of the outliers
# The positively skewed distibution of investment returns are
# generally more disired by investors

hsi['Skew50'] = hsi['Adj Close'].rolling(50).skew()
hsi.tail(3)
```

```python
# 67
# Drop the unnecessary columns
hsi=hsi.drop(['Open','High','Low','Close','Volume'],axis=1)
```

```python
hsi['250Max']=hsi['Adj Close'].rolling(250).max()
hsi['250Min']=hsi['Adj Close'].rolling(250).min()
hsi.tail(3)
```

```python
# 68
# Standard Deviation (250 days)
hsi['250Std'] = hsi['Adj Close'].rolling(250).std()
hsi.tail(3)
```

```python
# 69
# Value-at-risk (VaR) is a measure of the risk of loss of investment
# the lower 5% quantile at the risk is accepted
hsi['VaR'] = hsi['Adj Close'].rolling(250).quantile(0.05)
hsi.tail(3)
```

```python
# 70
# Aggregate rolling 250 median on 'Adj Close' and min on 'VaR'
# see first an example of .agg()
import pandas as pd
data={
    'AAA':[50,40,30],
    'BBB':[300,1112,42]
}
df=pd.DataFrame(data)
AAA=df.agg(['sum']) # ie AAA=50+40+30 and so does BBB
print(AAA)
```

```python
hsi['Adj Close'].rolling(250).agg({'Adj Close':'median','VaR':min}).tail()
```

```python
# 71
# Simplify HSI
hsi=hsi[['Adj Close']]
hsi.tail(3)
```

```python
# 72
# Import 10 yrs and 5 yrs treasury note data
# load 10 yrs US Treasury Note
tnx=yf.download('^TNX',start='2013-01-01',end='2023-01-01')
tnx=tnx[['Adj Close']]
tnx.tail(3)
```

```python
# load 5 yrs US Trasury Note

fvx=yf.download('^FVX',start='2013-01-01',end='2023-01-01')
fvx=fvx[['Adj Close']]
fvx.tail(3)
```

```python
# 73
# Merge 3 DFs on same Datetime index
# first see an example on .merge()
import pandas as pd

AAA={'Name':['jon','mary','dev'],'ID':[1,2,3],'Country':['hk','uk','us'],'Role':['ceo','cto','cfo']}
ADF=pd.DataFrame(AAA)
# print('ADF',ADF)
BDF=pd.DataFrame({'ID':[1,2,3],'Name':['jon','ann','leo']}) ; #print('BDF\n',BDF)
CM1=ADF.merge(BDF) ; print('merged\n',CM1)
CM2=ADF.merge(BDF,on='ID') ; print('CM2\n',CM2) # specific column
```

```python
df_all = pd.merge(pd.merge(tnx,fvx,on='Date'),hsi,on='Date')
df_all.head(3)
```

```python
# 74
# rename
df_all=df_all.rename(columns={'Adj Close_x':'10YrTN',
                             'Adj Close_y':'5YrTN',
                             'Adj Close':'HSI'})
df_all.head(3)
```

```python
# 75
# Difference of 10 and 5 Yrs tn
# see first an example of .diff()
# The .diff() method returns a DataFrame with the difference between
# the values for each row and, by default, the previous row
import pandas as pd
YY=pd.DataFrame({'A':[5,3,6,4],
                'B':[11,2,4,5],
                'C':[4,3,8,5],
                'D':[5,4,2,8]})
print(YY)
ZZ=YY.diff(axis=0, periods=1) ; print('ZZ\n',ZZ)
# row 1 - row 0 and row 2 - row 1 and so on
```

```python
diff = df_all[['5YrTN','10YrTN']].diff(axis=1)[['10YrTN']]
diff
```
